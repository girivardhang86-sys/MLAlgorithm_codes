# **This is kmeans algorithms**

```
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import numpy as np


# Generate synthetic data
X,y = make_blobs(n_samples=300,     # total points
                  centers=4,         # number of clusters
                  n_features=2,      # number of features (x, y)
                  cluster_std=1.5,   # spread of clusters
                  random_state=42)
# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Synthetic Clustered Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```
from sklearn.cluster import KMeans
wcss = []

for i in range(1,8):
    km = KMeans(n_clusters=i)
    y_means=km.fit_predict(X)
    wcss.append(km.inertia_)
```

# **train model of kmeans**

```

km = KMeans(n_clusters=3)
y_means = km.fit_predict(X)
```
